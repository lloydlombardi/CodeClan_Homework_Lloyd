Select the variables Beverage_category, Beverage, Beverage prep and Calories from the drinks_content data frame, and assign the selected columns to a new data frame called drinks. Check if there are any missing values anywhere in the data, and drop any rows containing them if there are.

```python
drinks = (
    drinks_content
    .loc[:, ["Beverage_category", "Beverage", "Beverage_prep", "Calories"]]
    .copy()
)

drinks
```

```python
drinks.info()
```

In your drinks dataframe create a new variable (column) called calorie_diff, which stores the difference between 135 calories (135 calories = 10 cubes of sugar!) and the calories in each drink.

```python
drinks = (
    drinks
    .assign(calorie_diff = 135 - drinks.Calories)
    .copy()
)

drinks
```

Summarise the mean number of calories in each beverage category.

```python
(
    drinks
    .groupby("Beverage_category")
    .Calories
    .mean()
)
```

Which Beverage_prep type contains more than the average calories of all drinks? 

```python
average_cals = (
    drinks
    .Calories
    .mean()
)

average_cals
```

```python
average_cals_prep = (
    drinks
    .groupby("Beverage_prep")
    .Calories
    .mean()
    .reset_index(name="mean_cals")
)

average_cals_prep
```

```python
(
    average_cals_prep
    .loc[average_cals_prep.mean_cals > average_cals, ["Beverage_prep", "mean_cals"]]
)
```

Which is the best type of coffee drink to get if you’re worried about consuming too many calories?

```python
drinks.Calories.sort_values().unique()
```

```python
(
    drinks
    .groupby("Beverage")
    .Calories
    .min()
    .sort_values()
)
```

Filter the dataset drinks so we only take "Classic Espresso Drinks", and save this in a new data frame called espresso_drinks.

```python
espresso_drinks = (
    drinks
    .loc[drinks.Beverage_category == "Classic Espresso Drinks"]
    .copy()
)

espresso_drinks
```

Group your espresso_drinks data frame by the type of Beverage prep, and then find out the mean calories for each drink group.

```python
(
    espresso_drinks
    .groupby("Beverage_prep")
    .Calories
    .mean()
)
```

Get the same grouped mean Calories values as above, but this time sorted in descending order. 

```python
espresso_mean_cals = (
    espresso_drinks
    .groupby("Beverage_prep")
    .Calories
    .mean()
    .reset_index(name="mean_cals")
)

espresso_mean_cals
```

```python
espresso_mean_cals.sort_values("mean_cals", ascending=False)
```

Visualisations

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
beverage_cals = (
    drinks
    .groupby("Beverage")
    .Calories
    .min()
    .sort_values()
    .reset_index(name="cals")
    .copy()
)
```

```python
sns.barplot(x="Beverage",
           y="cals",
           data=beverage_cals,
           color="steelblue")

plt.xticks(rotation = 45)

plt.show()
```

```python
espresso_mean_cals = (
    espresso_mean_cals
    .sort_values("mean_cals", ascending=False)
    .copy()
)
```

```python
sns.set_style("darkgrid")

sns.barplot(x="Beverage_prep",
           y="mean_cals",
           data=espresso_mean_cals,
           color="firebrick")

plt.xlabel("Beverage prep type")
plt.ylabel("Average calories")
plt.title("Average Calories of different Beverage Preps")


plt.xticks(rotation = 45)

plt.show()
```
