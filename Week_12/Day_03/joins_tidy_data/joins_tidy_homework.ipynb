Read in hat_observations and split observation into two columns, hat_colour and hat_type.

```python
hat = pd.read_csv('data/hat_observations.csv')
```

```python
hat
```

```python
split_df = hat.observation.str.split(pat=',', expand=True)
```

```python
hat['colour'] = split_df[0]
```

```python
hat['hat_style'] = split_df[1]
```

```python
hat = hat.drop(columns='observation')
hat
```

Unite day, month, and year columns into a column called date using a suitable separator. Then find the date where the most berets were observed.

```python
hat.info()
```

```python
hat.loc[:, ['day', 'month', 'year']] = hat.loc[:, ['day', 'month', 'year']].astype('string')
```

```python
hat.info()
```

```python
hat['date'] = hat[['day', 'month', 'year']].agg('-'.join, axis=1)
```

```python
hat
```

```python
(
    hat
    .loc[hat.hat_style == "beret", ['hat_style', 'observation_count', 'date']]
    .sort_values('observation_count', ascending=False)
    .head(1)
)
```

Can you join all 4 credit card datasets together so that you’re left with a dataset that looks like below with 109 rows and 12 columns?

```python
fraud_transaction = fraud_base.merge(transaction_base, how='left', on='transaction_id')

fraud_transaction
```

```python
fraud_transaction_card = fraud_transaction.merge(card_base, how='left', left_on='credit_card_id', right_on='card_number', copy=False)

fraud_transaction_card
```

```python
fraud_transaction_card_customer = fraud_transaction_card.merge(customer_base, how='left', on='cust_id')

fraud_transaction_card_customer
```

Read in exam_scores and transform it into long format with two new columns exam_question and score. Remove superfluous information from the values in exam_question.

```python
exam_scores = pd.read_csv('data/exam_scores.csv')
```

```python
exam_scores
```

```python
exam_scores = (
    exam_scores
        .rename(columns={
            'exam_Q1': 'Q1',
            'exam_Q2': 'Q2',
            'exam_Q3': 'Q3',
            'exam_Q4': 'Q4',
            'exam_Q5': 'Q5',
            'exam_Q6': 'Q6',
            'exam_Q7': 'Q7',
            'exam_Q8': 'Q8',
            'exam_Q9': 'Q9',
            'exam_Q10': 'Q10',
        })
)
```

```python
exam_scores
```

```python
exams_melted = (
    exam_scores
    .melt(
    id_vars=['id'],
    var_name='exam_question',
    value_name='score')
)
    
exams_melted
```
